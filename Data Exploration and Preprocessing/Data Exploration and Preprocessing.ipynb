### **Skewness Based On Its Type**
*   *Normally Distributed* : In this, the skewness is always equated to zero.  
          Skewness = 0
*  *Positively skewed distribution* : In this, A Positively-skewed distribution has a long right tail, thats why this is also known as right-skewed distribution. the reason behind it, in this value of mode is highest and mean is least which leads to right peak.
         Skewness > 0

  skew positif.PNG
*   *Negatively skewed distribution*: In this, a negatively skewed distribution has a long left tail, thats why this is also known as left-skewed distribution. the reason behind it, in this value of mode is least and mean is highest just reverse to right-skewed which leads to the left peak.
          Skewness < 0
 
 skew negatif.PNG



```python
# Menghitung skewness dari Income dan Age
# Import Library
from scipy.stats import skew

# Skewness Income
income_skewness = skew(df1['Income'])
sns.displot(df1['Income'],kind='kde')

skewness = ''
if(income_skewness > 0):
  skewness = 'Positively Skewed'
elif(income_skewness < 0):
  skewness = 'Negatively Skewed'
else:
  skewness = 'symmetrical'

print('Income Skewness : ', income_skewness, '(',skewness,')')
```

```python
# Skewness Age
age_skewness = skew(df1['Age'])
sns.displot(df1['Age'],kind='kde')

skewness = ''
if(income_skewness > 0):
  skewness = 'Positively Skewed'
elif(income_skewness < 0):
  skewness = 'Negatively Skewed'
else:
  skewness = 'symmetrical'

print('Age Skewness : ', age_skewness, '(',skewness,')')
```

*c. Hitung nilai minimum, maksimum, Q1, Q2, Q3 dari atribut Income dan Age.
Tentukan nilai inter quartile range (IQR) dari atribut Income dan Age.*

```python
# Max Income
max_income = df1['Income'].max()

# Min Income
min_income = df1['Income'].min()

# Q1, Q2, Q3 Income
q1_income = df1['Income'].quantile(q=0.25)
q2_income = df1['Income'].quantile(q=0.5)
q3_income = df1['Income'].quantile(q=0.75)

# Max Age
max_age = df1['Age'].max()

# Min Age
min_age = df1['Age'].min()

# Q1, Q2, Q3 Age
q1_age = df1['Age'].quantile(q=0.25)
q2_age= df1['Age'].quantile(q=0.5)
q3_age = df1['Age'].quantile(q=0.75)

#IQR Income
iqr_income = q3_income - q1_income

#IQR Age
iqr_age = q3_age - q1_age
```

*a. Lakukan smoothing dengan metode binning rata-rata pada atribute Age.
Gunakan bin berukuran 4 (bin depth). Buat grafik atribut Age sebelum dan
sesudah smoothing.*

```python
# Smoothing with Bin Mean for Age


data_age = df2['Age']

data_age = np.sort(data_age)

# Create 4 different bin matrices
bin1=np.zeros((10,5)) 


for i in range (0,50,5): 
    k=int(i/5) 
    mean=(data_age[i] + data_age[i+1] + data_age[i+2] + data_age[i+3] + data_age[i+4])/5
    for j in range(5): 
        bin1[k,j]=mean 

bin1
df2['binned_age'] = bin1.flatten()
```

g. Dalam sheet2 ada data yang nilainya tidak ada (missing value). Isi data
yang tidak ada tersebut dengan nilai yang sesuai.

Mengisi missing value pada data numerik dengan rata - rata masing -masing kolomnya. Dan mengisi missing value pada data kategorikal dengan data yang paling sering muncul

```python
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)

xt = DataFrameImputer().fit_transform(df2)
df2 = xt
df2
```

```python
genre = df2['Genre']
conditions = [(genre == 'Action'),
              (genre == 'Comedy'),
              (genre == 'Drama')]

choices = [0,1,2]
genre_encode = np.select(conditions, choices)
df2["Genre_Encode"] = genre_encode
```
